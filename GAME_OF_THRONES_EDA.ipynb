## **1.1 DATA DESCRIPTION**

In this Dataset, we have various information about each Episode/Season of Game of Thrones. The key features of this Dataset are - Season, No. of Episodes (Season), No. of Episodes (Overall), Title of the Episode, Running Time (Minutes), Directed by, Written by, Original Air Date, U.S. Viewers (Millions), Music by, Cinematography by, Editing by, IMDb Rating, Rotten Tomatoes Rating (Percentage), Metacritic Ratings, Season Ordered, Filming Duration, Novel(s) Adapted, Synopsis of each Episode.

Dataset Glossary (Column-wise)

* Season - No. of Seasons
* No. of Episode (Season) - No. of Episode in a particular Season
* No. of Episode (Overall) - No. of Episode in the whole Series
* Title of the Episode - Name of the Episode
* Running Time (Minutes) - Runtime of the Episode in Minutes
* Directed by - Name(s) of the Director
* Written by - Name(s) of the Writer
* Original Air Date - Original Air Date of the Episode
* U.S. Viewers (Millions) - No of U.S. Viewers of the Episode in Millions
* Music by - Name(s) of the Composer
* Cinematography by - Name(s) of the Cinematographer
* Editing by - Name(s) of the Editor
* IMDb Rating - IMDb Rating of the Episode (10 point scale)
* Rotten Tomatoes Rating (Percentage) - Rotten Tomatoes Rating of the Episode - in Percentage
* Metacritic Ratings - Metacritic Rating of the Episode
* Ordered - Date of Series/Season renewal
* Filming Duration - Filming Duration of the Season
* Novel(s) Adapted - Adapted from which Novel(s)
* Synopsis - Synopsis of the Episode

#### **HISTOGRAM AND DISPLOT**
A histogram is a bar graph-like representation of data that buckets a range of classes into columns along the horizontal x-axis. The vertical y-axis represents the number count or percentage of occurrences in the data for each column. Columns can be used to visualize patterns of data distributions.
<b>Hist</b> is used with the matplotlib library while <b>displot</b> is used with the seaborn library.


```python
sns.displot(GOT['Metacritic Ratings'], bins=10, kde=False, color = "m").set(title = "Metacritic Ratings displot")
plt.show()
```

The displot highlights the long left-handed tail for Metacritic Ratings. This confirms that the Metacritic Ratings is negatively skewed as seen with the box plot.

```python
sns.displot(GOT['Metacritic Ratings'], bins=10, kde=True, color = "m").set(title = "Metacritic Ratings displot")
plt.show()
```

Making the kde = True just add the density flow to it.

```python
plt.hist(GOT['Metacritic Ratings'], bins=20, color = "m")
plt.xlabel("Metacritic Ratings")
plt.ylabel("count")
plt.title ("Histogram of Metacritic Ratings")
plt.show()
```

The histogram highlights the long left-handed tail for Metacritic Ratings. This confirms that the Metacritic Ratings is negatively skewed as seen with the box plot.

```python
sns.displot(GOT['Season'], bins=10, kde=True, color = "m").set(title = "Season displot")
plt.show()
```

The displot highlights the right-handed tail for season 6 to 7. This shows that season 1 to 4 are evenly distributed while season 6 to 7 is positively skewed.

```python
sns.displot(GOT['U.S. Viewers (Millions)'], bins=10, kde=True, color = "m").set(title = "Viewers displot")
plt.show()
```

The histogram highlights the long right-handed tail for U.S. Viewers. This shows that the U.S. Viewers is positively skewed.

```python
sns.displot(GOT['IMDb Rating'], bins=10, kde=True, color = "y").set(title = "IMDb Rating displot")
plt.show()
```

The displot above shows IMDb Rating is negatively skewed.

**LINE PLOT**

Line graphs basically joins different points together. It is helpful for visualizing how a variable changes over time.

```python
x = GOT["IMDb Rating"]
y = GOT["Running Time (Minutes)"]
plt.plot(x, y, color = "#00ff59", linestyle = ":", marker = "d")
plt.xlabel("IMDb Rating")
plt.ylabel("Running Time (Minutes)")
plt.title("Lineplot of Running Time against IMDb Rating")
plt.show()
```

This shows the line plot between the Running Time and the IMDb Rating.

**KDE**

KDE stands for Kernel Density Estimator. A KDE plot gives us the sense of a univariate curve. The KDE plot below shows the skew distribution of <b>IMDb Rating</b>, <b>Metacritic Ratings</b>, <B>U.S. Viewers</b> and <b>Running Time</b>.

```python
sns.kdeplot(GOT['IMDb Rating'], shade=True)
sns.kdeplot(GOT['Metacritic Ratings'], shade=True)
sns.kdeplot(GOT['U.S. Viewers (Millions)'], shade=True)
sns.kdeplot(GOT['Running Time (Minutes)'], shade=True)

plt.legend(['IMDb Rating', 'Metacritic Ratings', 'U.S. Viewers (Millions)', 'Running Time (Minutes)'])
plt.xlabel(None)
plt.show()
```

The above shows the following;
*   Metacritic Ratings is negatively skewed
*   IMDb Rating is negatively skewed
*   Running Time (Minutes) is positively skewed
*   U.S. Viewers (Millions) is positively skewed

```python
corr = GOT.corr()
ones = np.ones_like(corr, dtype='bool')
mask = np.triu(ones)
mask = mask[1:, :-1]
corr = corr.iloc[1:,:-1].copy()
fig, ax = plt.subplots(figsize=(8,6))
sns.heatmap(corr, mask=mask, annot=True, fmt=".2f", cmap='PuBuGn', linewidths=.6, cbar_kws={'shrink':.9})
ax.xaxis.tick_bottom()
plt.title("Correlations heatmap for Game of Thrones dataset \n".upper(), fontdict={'fontsize': 18}, loc='left')
plt.show()
```
